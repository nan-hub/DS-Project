We need to get data using pandas datareader. We will get stock information from yahoo finance for the following banks:
*  Bank of America
* CitiGroup

* Morgan Stanley

**Figure out how to get the stock data from Jan 1st 2006 to Jan 1st 2020 for each of these banks. Set each bank to be a separate dataframe, with the variable name for that bank being its ticker symbol. This will involve a few steps:**
1. Use datetime to set start and end datetime objects.
2. Figure out the ticker symbol for each bank.
2. Figure out how to use datareader to grab info on the stock.
------------


```python
# Bank of America
BAC = data.DataReader(name = "BAC", data_source = "yahoo", start = "2006-01-01", end = "2020-01-01") 
BAC.head(3)
```

```python
# CitiGroup
C = data.DataReader(name = "C", data_source = "yahoo", start = "2006-01-01", end = "2020-01-01") 
C.head(3)
```

```python
# Goldman Sachs
GS = data.DataReader(name = "GS", data_source = "yahoo", start = "2006-01-01", end = "2020-01-01") 
GS.head(3)
```

```python
# JPMorgan Chase
JPM = data.DataReader(name = "JPM", data_source = "yahoo", start = "2006-01-01", end = "2020-01-01") 
JPM.tail(3)
```

```python
# Morgan Stanley
MS = data.DataReader(name = "MS", data_source = "yahoo", start = "2006-01-01", end = "2020-01-01") 
MS.head(3)
```

```python
# Wells Fargo
WFC = data.DataReader(name = "WFC", data_source = "yahoo", start = "2006-01-01", end = "2020-01-01") 
WFC.head(3)
```

**Create a list of the ticker symbols (as strings) in alphabetical order.**

```python
tickers = ["BAC", "C", "GS", "JPM", "MS", "WFC"]
```

**Use pd.concat to concatenate the bank dataframes together to a single data frame called bank_stocks. Set the keys argument equal to the tickers list.**

```python
bank_stocks = pd.concat([BAC, C, GS, JPM, MS, WFC], axis = 1, keys = tickers)
bank_stocks.head()
```

**Set the column name levels:**

```python
bank_stocks.columns.names = ['Bank Ticker','Stock Info']
```

**Check the head of the bank_stocks dataframe.**

```python
bank_stocks.head()
```

# EDA

Let's explore the data a bit!

**What is the max Close price for each bank's stock throughout the time period?**

```python
bank_stocks.xs(key = "Close", axis = 1, level = "Stock Info").max().round(2)
```

**Create a new empty DataFrame called returns. This dataframe will contain the returns for each bank's stock. returns are typically defined by:**

$$r_t = \frac{p_t - p_{t-1}}{p_{t-1}} = \frac{p_t}{p_{t-1}} - 1$$

```python
returns = pd.DataFrame()
```

**We can use pandas pct_change() method on the Close column to create a column representing this return value. Create a for loop that goes and for each Bank Stock Ticker creates this returns column and set's it as a column in the returns DataFrame.**

```python
for tick in tickers:
    returns[tick + ' Return'] = bank_stocks[tick]["Close"].pct_change()
returns.head()
```

**Create a pairplot using seaborn of the returns dataframe.**

```python
sns.set_style('whitegrid')
sns.pairplot(returns[1:])
```

We see the returns of Citizengroup stands out. We should check the background of Citizen's Stock Crash background to figure out what happened.

**Using this returns DataFrame, figure out on what dates each bank stock had the best and worst single day returns.**

```python
returns.idxmin()
```

We notice that 4 of the banks share the same day for the worst drop. 2009-01-20 in the Inauguration day.

```python
returns.idxmax()
```

**Get the standard deviation of the returns**

```python
returns.std() 
```

From the standard deviation, we may conclude that Citizengroup is the riskiest over the entire time period. 

```python
returns[returns.index.year == 2015].std() 
```

In the year of 2015, we may conclude that Bank of America or Morgan Stanley is riskest.

**Create a distplot using seaborn of the 2015 returns for Morgan Stanley**

```python
MS_Returns_2015 = returns[returns.index.year == 2015]["MS Return"]
```

```python
sns.set_style('whitegrid')
sns.distplot(MS_Returns_2015, bins = 100, color = "green")
```

**Create a distplot using seaborn of the 2008 returns for CitiGroup**

```python
C_Returns_2008 = returns[returns.index.year == 2008]["C Return"]
```

```python
sns.set_style('whitegrid')
sns.distplot(C_Returns_2008, bins = 100, color = "red")
```

## Moving Averages

Let's analyze the moving averages for these stocks in the year **2008**. 

**Plot the rolling 30 day average against the Close Price for Bank Of America's stock for the year 2008**

```python
plt.figure(figsize = (12, 6))
BAC[BAC.index.year == 2008].rolling(window = 30)["Close"].mean().plot(label = "30 Day Avg")
BAC[BAC.index.year == 2008]["Close"].plot(color = "green", label = "BAC CLOSE")
plt.legend()
```

**Create a heatmap of the correlation between the stocks Close Price.**

```python
close_corr = bank_stocks.xs(key = "Close", axis = 1, level = "Stock Info").corr()
sns.heatmap(close_corr, cmap = "coolwarm", annot = True)
```

**Use seaborn's clustermap to cluster the correlations together:**

```python
sns.clustermap(close_corr, cmap = "coolwarm", annot = True)
```

```python
close_corr.iplot(kind = "heatmap", colorscale = "rdylbu")
```
